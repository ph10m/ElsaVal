```python
import pandas as pd
import numpy as np
import os
import re
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
main_folder = os.path.dirname(os.getcwd())
source_path = os.path.join(main_folder, "strise", "STEP 2 - CSV_WITH_SENTIMENT", "entities.csv")
source_df = pd.read_csv(source_path)
source_df.drop(source_df.columns[[0, 1]], axis=1, inplace=True)
source_df
```

```python
eval_folder = os.path.dirname(os.getcwd())
eval_path = os.path.join(eval_folder, "strise", "STEP 2 - CSV_WITH_SENTIMENT (eval)", "entities.csv")
eval_df = pd.read_csv(eval_path)
eval_df.drop(eval_df.columns[[0, 1]], axis=1, inplace=True)
eval_df
```

```python
source_ids = set(source_df.id)
eval_ids = set(eval_df.id)
```

```python
valid_ids = []
for eval_id in eval_ids:
    if eval_id not in source_ids:
        valid_ids.append(eval_id)
```
