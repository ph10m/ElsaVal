```python
jsonfile = "../graphql/downloads/sports/2020_Q1.json"
rels = set()
from collections import defaultdict
countrels = defaultdict(int)

instance_of_inspector = set()

with open(jsonfile, 'r', encoding="utf8") as f:
    f = json.load(f)
    edges = f["events"]["edges"]

    entities = []
    texts = []

    for node in edges:
        node = node["node"]
        #print(node)
        for egs in node["entities"]:
            for ent in node["entities"][egs]:
                try:
                    tmp_rels = ent["node"]["relations"]["edges"]
                    for tr in tmp_rels:
                        _typ = tr["relationship"]["name"]
                        if "instance_of" in _typ:
                            #rels.add(tr["node"]["name"])
                            typname = tr["node"]["name"]
                            countrels[typname] += 1
                            
                            if "activity" in typname.lower():
                                instance_of_inspector.add(ent["node"]["name"])
                    
                except:
                    pass
                        
        
for k,v in sorted(countrels.items(), key=lambda x: -x[1]):
    print(k,v)
```
